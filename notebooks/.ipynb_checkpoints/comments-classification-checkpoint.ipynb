Предположим от продакт менеджера поступила задача разработать систему которая будет банить плохие комментарии.
Нужно разработать бинарный классификатор который на вход будет принимать текст комментария и на выходде выдавать 2 класса (хороший/плохой комментарий). Пусть у нас задача максимизировать количество удалённых негативных комментариев, но при этом не стоит удалять все комментарии, дабе не зацепить лишние и не попортить юзер экспириенс

Наша задача максимизировать recall и при этом precision >= 0.95

Датасет для такой задачи можно собрать вручную путём самсотятельной сборки и разметки комментариев это может оказаться достаточно долгим и дорогим

### **Предобработка данных**

Перед тем как подавать данные на вход модели надо из текста сделать численные векторы

Для этого нужно:
- Разбить текст на токены
- Удалить токены, не несущие смысловой инфы
- Удалить стоп-слова (междометия, вводыне слова и т.д)
- У слова надо будет удалить окончания
- Привести все слова к единимоу регистру


```python
sentence_example = df.iloc[1]['comment']
tokens = word_tokenize(sentence_example, language='russian')
russian_stop_words = stopwords.words('russian')
tokens_without_punctuation_and_stop_words = [i for i in tokens if i not in string.punctuation and i not in russian_stop_words]
snowball = SnowballStemmer(language='russian')
stemmed_tokens = [snowball.stem(i) for i in tokens_without_punctuation_and_stop_words]
```

```python
print(f'Исходный текст: {sentence_example}')
```

```python
print(f'Токены: {tokens}')
```

```python
print(f'Токены без пунктуации и стоп-слов: {tokens_without_punctuation_and_stop_words}')
```

```python
print(f'Токены после стемминга: {stemmed_tokens}')
```

Далее по тому же принципу напишем функцию, позволяющую обработать весь датасет

```python
snowball = SnowballStemmer(language='russian')
russian_stop_words = stopwords.words('russian')
def tokenize_sentence(sentence: str, remove_stop_words: bool = True):
    tokens = word_tokenize(sentence, language='russian')
    tokens = [i for i in tokens if i not in string.punctuation]
    if remove_stop_words:
        tokens = [i for i in tokens if i not in russian_stop_words]
    tokens = [snowball.stem(i) for i in tokens]
    return tokens
```

```python
tokenize_sentence(df.iloc[1]['comment'])
```

Механизм подготовки массива токенов готов, можно применять TF-IDF векторизацию для кодирования последовательности токенов


```python
vectorizer = TfidfVectorizer(
    tokenizer=lambda x: tokenize_sentence(x, remove_stop_words=True),
    max_features=10000,
    min_df=2,
    max_df=0.95,
    token_pattern=None)
```

### **Обучение**

```python
features = vectorizer.fit_transform(train_df['comment'])
```

```python
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(features, train_df['toxic'])
```

```python
print(f'Предсказание модели для коммента: {model.predict(features[0])} \nСам коммент: {train_df['comment'].iloc[0]}')
```

```python
model_pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer(tokenizer=lambda x: tokenize_sentence(x, remove_stop_words=True), token_pattern=None)),
    ('model', LogisticRegression(class_weight='balanced', random_state=42))
])
```

```python
model_pipeline.fit(train_df['comment'], train_df['toxic'])
```

**Проверим:**

```python
model_pipeline.predict(['А чувствуется разница между бомбером и, допустим, этим концертом?'])
```

```python
model_pipeline.predict(['Шо то хуйня, шо это хуйня. Вот это обе хуйни такие, шо я бля ебал её маму рот'])
```

### **Оценка качества**

```python
precision_score(y_true=test_df['toxic'], y_pred=model_pipeline.predict(test_df['comment']))
```

```python
recall_score(y_true=test_df['toxic'], y_pred=model_pipeline.predict(test_df['comment']))
```

```python
precision, recall, thresholds = precision_recall_curve(
    y_true=test_df['toxic'],
    y_score=model_pipeline.predict_proba(test_df['comment'])[:, 1]
)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='PR Кривая')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Кривая')
plt.grid(True)
plt.legend()
plt.show()
```

```python
np.where(precision > 0.95)
```

```python
thresholds[397]
```

```python
precision_score(y_true=test_df['toxic'], y_pred=model_pipeline.predict_proba(test_df['comment'])[:, 1] > thresholds[397])
```

Получено приемлемое значение precision, модель почти не ошибается в "обвинениях"

Посчитаем recall при таком значении

```python
recall_score(y_true=test_df['toxic'], y_pred=model_pipeline.predict_proba(test_df['comment'])[:, 1] > thresholds[397])
```

Получилась модель осторожная, токсичность предсказывается когда модель оч уверена в себе

Могут пропускаться реальыне токсики из-за того что модель не решилась отметить их таковыми

В проде такая модель не испортит юзер экспириенс забанив добропорядчных комментаторов, но в то же время может не оправдать ожиданий, пропустив нарушителей

### **Подбор гиперпараметров**

```python
grid_pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer(tokenizer=lambda x: tokenize_sentence(x, remove_stop_words=True), max_features=10000, min_df=2, max_df=0.95, token_pattern=None)),
    ('model', 
     GridSearchCV(
         LogisticRegression(class_weight='balanced', random_state=0),
         param_grid={
             'C': [0.1, 1, 10.],
            'penalty': ['l2'],
            'solver': ['lbfgs', 'liblinear'],
            'class_weight': [None, 'balanced']
         },
         cv=3,
         verbose=4
         )
     )
])
```

```python
grid_pipeline.fit(train_df['comment'], train_df['toxic'])
```

```python
model_pipeline_c_10 = Pipeline([
    ('vectorizer', TfidfVectorizer(tokenizer=lambda x: tokenize_sentence(x, remove_stop_words=True), token_pattern=None)),
    ('model', LogisticRegression(class_weight='balanced', random_state=42, C=10.))
])
```

```python
model_pipeline_c_10.fit(train_df['comment'], train_df['toxic'])
```

```python
precision_c_10, recall_c_10, thresholds_c_10 = precision_recall_curve(
    y_true=test_df['toxic'],
    y_score=model_pipeline_c_10.predict_proba(test_df['comment'])[:, 1]
)
```
